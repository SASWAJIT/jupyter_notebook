```python
import matplotlib.pyplot as plt
x=['Day1','Day2','Day3','Day4','Day5']
y=[200,450,400,520,470]
y1=[270,300,400,380,430]
plt.plot(x,y,marker='^',ls=':',color='Red',label='Week1')
plt.plot(x,y1,marker='*',ls='-',color='Green',label='Week2',alpha=0.4)
plt.legend()
plt.show()
```
