```python
import matplotlib.pyplot as plt
import numpy as np
mobile=['Apple','samsung','Oneplus','Nokia','Redmi']
x=[22,35,30,3,10]
ex=[0,0,0,0,0.1]
plt.pie(x,labels=mobile,explode=ex,shadow='True',autopct='%.f',startangle=180)
plt.show()
```
