```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,20,15,14,22]
plt.subplot(2,2,1)
plt.title("Age")
plt.bar(x,y)

x=[5,6,7,8,9]
y=[14,20,16,17,21]
plt.subplot(2,2,2)
plt.title("Weight")
plt.suptitle("Employee Data")
plt.plot(x,y)
plt.savefig("Sub_plt2.png",facecolor='green',bbox_inches='tight',pad_inches=0.4)
plt.show()
```
